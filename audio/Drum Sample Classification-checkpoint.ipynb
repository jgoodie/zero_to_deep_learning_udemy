```python
def get_features(file_name):
    X, sr = librosa.load(file_name)
    stft = np.abs(librosa.stft(X))
    mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sr, n_mfcc=40).T,axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sr).T,axis=0)
    mel = np.mean(librosa.feature.melspectrogram(X, sr=sr).T,axis=0)
    contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sr).T,axis=0)
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(X),sr=sr).T,axis=0)
    return mfccs,chroma,mel,contrast,tonnetz
```

```python
features, labels = np.empty((0,193)), np.empty(0)
mfccs, chroma, mel, contrast, tonnetz = get_features('samples/Floor/93_SwngFnky_FloorTom.wav')
ext_features = np.hstack([mfccs,chroma,mel,contrast,tonnetz])
features = np.vstack([features,ext_features])
```

```python
pd.DataFrame(features)
```

```python
def get_sample_vectors(path):
    class_label = []
    features, labels = np.empty((0,193)), np.empty(0)
    for x in glob.glob(path):
        class_label.append(x.split('/')[1])
        mfccs, chroma, mel, contrast, tonnetz = get_features(x)
        ext_features = np.hstack([mfccs,chroma,mel,contrast,tonnetz])
        features = np.vstack([features,ext_features])
    return features, class_label
```

```python
df['class'] = c
```

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
y_cat = pd.get_dummies(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.33, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
K.clear_session()
model = Sequential()
model.add(Dense(32, input_shape=(193,), kernel_initializer='lecun_uniform', activation='relu'))
model.add(Dense(16, kernel_initializer='lecun_uniform', activation='relu'))
model.add(Dense(8, kernel_initializer='lecun_uniform', activation='relu'))
model.add(Dense(9, activation='softmax'))

model.compile(RMSprop(lr=0.001),'categorical_crossentropy', metrics=['accuracy'])

model.fit(np.array(X_train), np.array(y_train.values), batch_size=16, epochs=200, verbose=1)
```

```python
model.summary()
```

```python
model.evaluate(np.array(X_train), np.array(y_train))
```

```python
model.evaluate(np.array(X_test), np.array(y_test))
```

```python
y_pred = model.predict_classes(np.array(X_test))
```
